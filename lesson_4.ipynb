```python
# Задача 4. Декоратор для измерения времени выполнения
# Реализуйте декоратор timer

import time
import functools

# Декоратор timer
def timer(func):
    # 5. Сохраняет имя (__name__) и документацию (__doc__) исходной функции
    @functools.wraps(func)
    # 4. Корректно работает с любыми функциями (любое количество аргументов)
    def wrapper(*args, **kwargs):
        # 1. Замеряет время начала декорируемой функции
        start_time = time.time()
        
        result = func(*args, **kwargs)
        
        # 1. Замеряет время окончания декорируемой функции
        end_time = time.time()
        
        # 2. Выводит на экран сообщение: "Функция {имя_функции} выполнилась за {время:.4f} сек."
        execution_time = end_time - start_time
        print(f"Функция {func.__name__} выполнилась за {execution_time:.4f} сек.")
        
        # 3. Возвращает результат выполнения функции
        return result

    return wrapper


@timer
def sum_to_n(n=1_000_000):
    """Считает сумму чисел."""
    return sum(range(1, n + 1))

result = sum_to_n(1_000_000)

print(f"Результат: {result}")

# Для Задачи 4 реализовано сохранение метаданных
print(f"Имя функции: {sum_to_n.__name__}")
print(f"Документация: {sum_to_n.__doc__}")
```
